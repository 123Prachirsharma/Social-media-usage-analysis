```python
import pandas as pd 
import numpy as np

# Load the dataset
file_path = "D:/python project 2/Social Media Usage India.csv"
data = pd.read_csv(file_path)
data.head(10)
```

```python
print(data.isnull().sum())
```

```python
cols_to_clean = [
    'Time Spent on Facebook in last week (in minutes)', 
    'Time Spent on Facebook in last weekend (in minutes)', 
    'Time Spent on WhatsApp in last weekend (in minutes)', 
    'Total Instagram Usage', 
    'Total Week Usage', 
    'Total WhatsApp Usage'
]

# Remove all characters except digits, decimal point, and comma
for col in cols_to_clean:
    data[col] = data[col].astype(str).str.replace(r'[^\d.,]', '', regex=True) 

# The cleaned data remains in the DataFrame `data`
data
```

```python

# Function to remove commas from string values only
def remove_commas(val):
    if isinstance(val, str):  # Only apply if the value is a string
        return val.replace(',', '')
    return val  # Return the value unchanged if it's not a string

# Apply the function to the relevant columns
columns_to_clean = [
    'How many followers do you have on Instagram? (In case of multiple accounts, please mention the one with the maximum)',
    'How many posts do you have on Instagram?',
    'Time Spent on Facebook in last week (in minutes)',
    'Time Spent on Facebook in last weekend (in minutes)',
    'Time Spent on Instagram in last week (in minutes)',
    'Time Spent on Instagram in last weekend (in minutes)',
    'Time Spent on WhatsApp in last week (in minutes)',
    'Time Spent on WhatsApp in last weekend (in minutes)',
    'Total Facebook Usage',
    'Total Instagram Usage',
    'Total Social Media Usage',
    'Total Week Usage',
    'Total Weekend Usage',
    'Total WhatsApp Usage'
]

for column in columns_to_clean:
    data[column] = data[column].apply(remove_commas)

# Display the cleaned DataFrame
print(data)
```

```python
df = pd.DataFrame(data)

# Display the datatype of each column
print(df.dtypes)
```

```python
df['How many followers do you have on Instagram? (In case of multiple accounts, please mention the one with the maximum)'] = pd.to_numeric(df['How many followers do you have on Instagram? (In case of multiple accounts, please mention the one with the maximum)'], errors='coerce')
df['How many posts do you have on Instagram?'] = pd.to_numeric(df['How many posts do you have on Instagram?'], errors='coerce')
df['Time Spent on Facebook in last week (in minutes)'] = pd.to_numeric(df['Time Spent on Facebook in last week (in minutes)'], errors='coerce')
df['Time Spent on Facebook in last weekend (in minutes)'] = pd.to_numeric(df['Time Spent on Facebook in last weekend (in minutes)'], errors='coerce')
df['Time Spent on Instagram in last week (in minutes)'] = pd.to_numeric(df['Time Spent on Instagram in last week (in minutes)'], errors='coerce')
df['Time Spent on Instagram in last weekend (in minutes)'] = pd.to_numeric(df['Time Spent on Instagram in last weekend (in minutes)'], errors='coerce')
df['Time Spent on WhatsApp in last week (in minutes)'] = pd.to_numeric(df['Time Spent on WhatsApp in last week (in minutes)'], errors='coerce')
df['Time Spent on WhatsApp in last weekend (in minutes)'] = pd.to_numeric(df['Time Spent on WhatsApp in last weekend (in minutes)'], errors='coerce')
df['Total Facebook Usage'] = pd.to_numeric(df['Total Facebook Usage'], errors='coerce')
df['Total Instagram Usage'] = pd.to_numeric(df['Total Instagram Usage'], errors='coerce')
df['Total Social Media Usage'] = pd.to_numeric(df['Total Social Media Usage'], errors='coerce')
df['Total Week Usage'] = pd.to_numeric(df['Total Week Usage'], errors='coerce')
df['Total Weekend Usage'] = pd.to_numeric(df['Total Weekend Usage'], errors='coerce')
df['Total WhatsApp Usage'] = pd.to_numeric(df['Total WhatsApp Usage'], errors='coerce')

# For categorical columns (like gender, city, etc.), we don't need to convert the data type since they should remain as 'object' type.
df['Age'] = df['Age'].astype('int64')
df['Latitude'] = df['Latitude'].astype('float64')
df['Longitude'] = df['Longitude'].astype('float64')

# Check the new data types
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
# Replace empty strings with NaN in the columns to clean
cols_to_clean = [
    'Time Spent on WhatsApp in last weekend (in minutes)',
    'Time Spent on WhatsApp in last week (in minutes)',
    'Time Spent on Instagram in last weekend (in minutes)',
    'Time Spent on Facebook in last weekend (in minutes)',
    'Total WhatsApp Usage',
    'Total Weekend Usage'
]

for col in cols_to_clean:
    df[col] = df[col].replace('', np.nan)  # Replace empty strings with NaN

# Filling 'Total WhatsApp Usage' (Column Z)
df['Total WhatsApp Usage'] = (
    df['Time Spent on WhatsApp in last weekend (in minutes)'].fillna(0).astype(float) +
    df['Time Spent on WhatsApp in last week (in minutes)'].fillna(0).astype(float)
)

# Filling 'Total Weekend Usage' (Column Y)
df['Total Weekend Usage'] = (
    df['Time Spent on WhatsApp in last weekend (in minutes)'].fillna(0).astype(float) +
    df['Time Spent on Instagram in last weekend (in minutes)'].fillna(0).astype(float) +
    df['Time Spent on Facebook in last weekend (in minutes)'].fillna(0).astype(float)
)

# Filling 'Time Spent on WhatsApp in last weekend (in minutes)' (Column T)
df['Time Spent on WhatsApp in last weekend (in minutes)'] = (
    df['Total WhatsApp Usage'].fillna(0).astype(float) -
    df['Time Spent on WhatsApp in last week (in minutes)'].fillna(0).astype(float)
)

# Filling 'Time Spent on WhatsApp in last week (in minutes)' (Column S)
df['Time Spent on WhatsApp in last week (in minutes)'] = (
    df['Total WhatsApp Usage'].fillna(0).astype(float) -
    df['Time Spent on WhatsApp in last weekend (in minutes)'].fillna(0).astype(float)
)

# Filling 'Time Spent on Instagram in last weekend (in minutes)' (Column R)
df['Time Spent on Instagram in last weekend (in minutes)'] = (
    df['Total Weekend Usage'].fillna(0).astype(float) -
    df['Time Spent on WhatsApp in last weekend (in minutes)'].fillna(0).astype(float) -
    df['Time Spent on Facebook in last weekend (in minutes)'].fillna(0).astype(float)
)

# Save the updated DataFrame to a CSV file
df.to_csv("updated_data_new.csv", index=False)

print("The updated data has been saved to 'updated_new_n.csv'.")
```

## Q5. Do users with high Instagram followers also spend significant time on Instagram?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "D:/python project 2/updated_data_new.csv"  # Replace with your actual file path
data_frame = pd.read_csv(file_path)

# Define bins for categorizing follower counts
bins = [0, 100, 1000, 10000, 100000, 1000000]
labels = ['0-100', '101-1K', '1K-10K', '10K-100K', '100K+']
followers_column = 'How many followers do you have on Instagram? (In case of multiple accounts, please mention the one with the maximum)'

# Categorize follower counts
data_frame['Follower Category'] = pd.cut(data_frame[followers_column], bins=bins, labels=labels, right=False)

# Group by follower category and calculate average time spent on Instagram
avg_time_spent = data_frame.groupby('Follower Category')[
    ['Time Spent on Instagram in last week (in minutes)', 
     'Time Spent on Instagram in last weekend (in minutes)']
].mean().reset_index()

# Plot the results
fig, ax = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Weekly time spent
ax[0].bar(avg_time_spent['Follower Category'], avg_time_spent['Time Spent on Instagram in last week (in minutes)'], color='skyblue')
ax[0].set_title('Average Weekly Time Spent on Instagram')
ax[0].set_ylabel('Time (minutes)')
ax[0].set_xlabel('Follower Category')

# Weekend time spent
ax[1].bar(avg_time_spent['Follower Category'], avg_time_spent['Time Spent on Instagram in last weekend (in minutes)'], color='lightgreen')
ax[1].set_title('Average Weekend Time Spent on Instagram')
ax[1].set_xlabel('Follower Category')

plt.tight_layout()
plt.show()
```

## Q1.Who are the top 10% of users by total social media usage?

```python
top_10_percent_threshold = databases["Total Social Media Usage"].quantile(0.9)
top_10_percent_users = databases[databases["Total Social Media Usage"] > top_10_percent_threshold]

plt.figure(figsize=(8, 6))
sns.histplot(top_10_percent_users["Total Social Media Usage"], kde=True, bins=20, color='skyblue')
plt.title("Distribution of Total Social Media Usage (Top 10% Users)", fontsize=16)
plt.xlabel("Total Social Media Usage (minutes)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

## Q2.Are high usage outliers concentrated in specific geographic regions or cities?

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=top_10_percent_users, x="City", order=top_10_percent_users["City"].value_counts().index, palette="viridis")
plt.title("Geographic Concentration of High-Usage Outliers", fontsize=16)
plt.xlabel("City", fontsize=12)
plt.ylabel("Count of High-Usage Outliers", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

## Q3.What is the average age and gender distribution of high usage outliers?

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=top_10_percent_users, x="Gender", y="Age", palette="Set2")
plt.title("Age Distribution of High-Usage Outliers by Gender", fontsize=16)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Age", fontsize=12)
plt.show()
```

## Q4.Which platforms contribute the most to high total usage values?

```python
platform_contributions = top_10_percent_users[
    ["Total Facebook Usage", "Total Instagram Usage", "Total WhatsApp Usage"]
].mean()
plt.figure(figsize=(8, 6))
platform_contributions.plot(kind='bar', color=['#4C72B0', '#55A868', '#C44E52'])
plt.title("Platform Contributions to High Total Usage", fontsize=16)
plt.ylabel("Average Time Spent (minutes)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```
